```python
import pandas as pd
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv('data.csv')
```

```python
df = df.dropna()
```

```python
df = df.drop_duplicates(subset=['id'], keep='last')
```

```python
df
```

```python
df.dtypes
```

```python
data = pd.read_excel('IFND_dataset.xlsx')
data
```

```python
final_data = pd.merge(df, data, how="left", on="id")
final_data
```

```python
final_data['Label'] = final_data['Label'].map({
    True : 1,
    'Fake' : 0
})
final_data
```

```python
final_data['Pred1'] = final_data.apply(lambda x: (1 if x['entailment'] > x['contradiction'] else 0), axis=1)
final_data
```

```python
print(classification_report(y_true=final_data['Label'], y_pred=final_data['Pred1'], zero_division=0))
```

```python
final_data['Pred2'] = final_data.apply(lambda x: (1 if (x['entailment'] + x['neutral'] > x['contradiction']) else 0), axis=1)
final_data
```

```python
print(classification_report(y_true=final_data['Label'], y_pred=final_data['Pred2'], zero_division=0))
```

```python
accuracy1 = (final_data['Label'] == final_data['Pred1']).mean() * 100
accuracy1
```

```python
accuracy2 = (final_data['Label'] == final_data['Pred2']).mean() * 100
accuracy2
```

```python
def classify(x):
    if(x['neutral'] + x['contradiction'] > x['contradiction']):
        return 1
    elif(x['contradiction'] > x['entailment']):
        return 0
    else:
        return 0
```

```python
final_data['Pred3'] = final_data.apply(classify, axis=1)
final_data
```

```python
accuracy3 = (final_data['Label'] == final_data['Pred3']).mean() * 100
accuracy3
```
