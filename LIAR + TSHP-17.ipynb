```python
import numpy as np
import pandas as pd
```

## Reading & Merging Datasets

```python
columns_LIAR = ['id','label','text','subject','speaker','job title','state info','party','barely true','false','half true','mostly true','pants on fire','context']
columns_TSHP = ['speaker', 'label', 'text', 'explanation', 'url']
```

```python
LIAR_train = pd.read_table("LIAR.tsv", sep="\t", header= None, names=columns_LIAR)
print(LIAR_train.shape)
LIAR_train.head()
```

```python
LIAR_valid = pd.read_table("LIAR_VALID.tsv", sep="\t", header= None, names=columns_LIAR)
print(LIAR_valid.shape)
LIAR_valid.head()
```

```python
LIAR_test = pd.read_table("LIAR_TEST.tsv", sep="\t", header= None, names=columns_LIAR)
print(LIAR_test.shape)
LIAR_test.head()
```

```python
TSHP = pd.read_table("TSHP_17.csv", sep="\t", header= None, skiprows=(0,1,2,4021,6315,9380))
TSHP = TSHP.drop([5], axis=1)
TSHP.columns = columns_TSHP
```

```python
print(TSHP.shape)
TSHP.head()
```

```python
TSHP_label_map = {5:'pants-fire', 4:'false', 3:'barely-true', 2:'half-true', 1:'mostly-true', 0:'true'}
```

```python
TSHP['label'] = TSHP['label'].map(TSHP_label_map)
```

```python
TSHP.head()
```

```python
LIAR = pd.concat([LIAR_train, LIAR_valid, LIAR_test], ignore_index=True)
```

```python
print(LIAR.shape)
LIAR.head()
```

```python
LIAR
```

```python
LIAR_select = LIAR[['text', 'speaker','label']]
print(LIAR_select.shape)
LIAR_select.head()
```

```python
TSHP_select = TSHP[['text', 'speaker','label']]
print(TSHP_select.shape)
TSHP_select.head()
```

```python
data = pd.concat([LIAR_select, TSHP_select], ignore_index=True)
data
```

```python
data = data.drop_duplicates('text', keep='last', ignore_index=True)
```
